```python
import pandas as pd
import os
import cv2
```

```python
n = 0
videos = os.listdir('../../data/trafficdb/video/')

while n != 253:
    i = 0
    video_name = videos[n]
    cap = cv2.VideoCapture('../../data/trafficdb/video/' + video_name)
    
    dir_name = '../../data/frames/' + video_name[0:24] + '/'
    if not os.path.exists(dir_name):
        os.makedirs(dir_name)
    
    while(cap.isOpened()):
        ret, frame = cap.read()
        if ret:
            cv2.imwrite(dir_name + 'frame' + str(i) + '.png', frame)
            i += 1
        else:
            break
    print video_name + ' is divided on frames'        
    n += 1
    cv2.destroyAllWindows()
print 'Complete!'
```
